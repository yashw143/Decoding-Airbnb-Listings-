````python
# Chart - 5 visualization code
neighbourhood_prices = df.groupby('neighbourhood')['price'].mean().reset_index()
# top_10_neighbourhood.sort_values(by='price',ascending=False)

#Renam coluumns
neighbourhood_prices.columns =['neighbourhood','Average_price_per_neighborhood']

#sort in ascneding order and selecct the top10 neighbourhoods
top_10_neighbourhood = neighbourhood_prices.sort_values('Average_price_per_neighborhood',ascending=False).head(10)

#Rest the index and rename the row names
# top_10_neighbourhood = top_10_neighbourhood.rest_index(drop=True)


top_10_neighbourhood  = top_10_neighbourhood.reset_index(drop=True)
top_10_neighbourhood.index = top_10_neighbourhood.index+1

# top_10_neighbourhood

least_10_neighbourhood = neighbourhood_prices.sort_values('Average_price_per_neighborhood',ascending=False).tail(10)
least_10_neighbourhood = least_10_neighbourhood.reset_index(drop=True)
least_10_neighbourhood.index = least_10_neighbourhood.index+1
````

This chart provides insights into the average price of Airbnb listings for the top 10 neighborhoods. The bar graph shows the average price per neighborhood, and the text labels on each bar provide specific information about the average price for each neighborhood.

````python
plt.rcParams["figure.figsize"] = (20,9)
neighbourhood = top_10_neighbourhood['neighbourhood']
average = top_10_neighbourhood['Average_price_per_neighborhood']
my_colors = ['#F6B5C3', '#D8E5E7', '#F9C7A9', '#B8F2E7', '#D1C4E9', '#A3D9E8', '#B6F4F2', '#A6D9C1', '#F5B2A7', '#C8E0A7']
plt.bar(neighbourhood,average,color=my_colors)
plt.title("Top 10 neighbourhood by the average price",fontsize=18,fontweight='bold')
plt.xlabel("The specific neighborhood or district where the property is located",fontsize=18,fontweight='bold')
plt.xticks(fontsize=16,fontweight='bold',rotation=85,color='red')
plt.yticks(fontsize=13,fontweight='bold')
# Add the values to each bar
for i,v in enumerate(round(average,2)):
  plt.text(i,v//2,str(v),color='black',ha='center',fontweight='bold',fontsize=14,bbox=dict(facecolor='orange', alpha=0.5))
plt.tight_layout()
````
